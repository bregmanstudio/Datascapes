```python
# A simple computer music instrument (CMI)
epd = 8 # events per day
evy = epd*365.25 # events per year
ev_map = {'d':epd,'m':evy/12,'12m':evy,'w':epd*7} # events per time-period

modes = {
    'major': [0, 2, 4, 5, 7, 9, 11],
    'major_bass': [0, 7, 2, 9, 4, 11, 5],
    'minor': [0, 2, 3, 5, 7, 8, 11],
    'minor_bass': [0, 7, 2, 8, 3, 11, 5]
}

def make_note(note, vol, dur, base_freq=440.0, **params):
    params.setdefault('num_harmonics',10)
    params.setdefault('sample_rate',44100) 
    params.setdefault('num_points', int(round(dur*params['sample_rate'])))
    w = hanning(88)
    wn = len(w) // 2.0
    h = vol * harmonics(f0=base_freq*2**(note/12.0), **params) # 
    h[:wn] = h[:wn] * w[:wn]
    h[-wn:] = h[-wn:] * w[-wn:]
    return h

def mode_map(n, mode=modes['major']):    
    n = int(round(n)) 
    pc = n%len(mode)
    octave = n//len(mode)
    return mode[pc] + octave*12

def map_pitch(x, rule='12m', how='mean', mode=modes['major']):
    x2 = x.resample(rule, how, closed='left')
    for i, v in enumerate(x2['value']):
        x2['value'][i] = mode_map(v, mode)
    return x2

def synthesize(x, dur=1.0, f0=220.0, **params):
    snd = [make_note(n, 1.0, dur, base_freq=f0, **params) for n in x['value']]
    return balance_signal(hstack(snd))
```

```python
res2 = res.copy()
res2 *= 3
y_notes = map_pitch(res2, '12m','median', modes['major_bass'])
for i, yr in enumerate(y_notes.index.year):
    res2['{}'.format(yr)]['value'] = res2['{}'.format(yr)]['value']  + y_notes['value'][i] % 12
m_notes = map_pitch(res2, 'm','median', modes['major'])
w_notes = map_pitch(res2, 'w','median', modes['major'])
d_notes = map_pitch(res2, 'd','median', modes['major'])
```

```python
y_snd = synthesize(y_notes, 4.0, 27.5, afun=lambda a: a)
m_snd = synthesize(m_notes, 4.0*ev_map['m']/evy, 110.)
w_snd = synthesize(w_notes, 4.0*ev_map['w']/evy, 440, num_harmonics=1)
d_snd = synthesize(d_notes, 4.0*ev_map['d']/evy, 1760, num_harmonics=1)
```

```python
snd_lens = len(y_snd), len(m_snd), len(w_snd), len(d_snd)
print snd_lens 
n = min(snd_lens)
y_snd = y_snd[:n]
m_snd = m_snd[:n]
w_snd = w_snd[:n]
d_snd = d_snd[:n]
print "nsamples = ", n, "=", round(n/44100.,2), "secs."
```

```python
amp = 4
play(balance_signal(amp**y_snd + m_snd + (amp**-1)*w_snd + (amp**-2)*d_snd))
```
